## merge df

```python
mdf1=pd.DataFrame({'id':[1,2,3,4],'name':['A','B','C','D']})
```

```python
mdf2=pd.DataFrame({'id':[1,2,3,4],'marks':['88','92','79','100']})
```

```python
#on should take common col in both df
pd.merge(mdf1,mdf2,on='id')
```

```python
mdf2=pd.DataFrame({'id':[1,2,3,5],'marks':['88','92','79','100']})
```

```python
#only common values in id col are merged i.e. 1,2,3 
#similar to inner join
pd.merge(mdf1,mdf2,on='id')
```

```python
#left join
pd.merge(mdf1,mdf2,on='id',how='left')
```

```python
#outer join
pd.merge(mdf1,mdf2,on='id',how='outer')
```

```python
#indicator
pd.merge(mdf1,mdf2,on='id',how='outer',indicator=True)
```

```python
mdf3=pd.DataFrame({'id':[5,6,7,8],'marks':['88','92','79','100']})
```

```python
#used when no common values to join
pd.merge(mdf1,mdf3,left_index=True,right_index=True)
```
